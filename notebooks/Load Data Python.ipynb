## Load the data with the funtion open

```python
#first parameter is the path from where you are going to read the data
#secon parameter 'r' is to read only , 'rw' is to read and write and 'w' is to write
#we can use the 'a' if we want to write include more information in the file 
df3 = open("../datasets/customer-churn-model/Customer Churn Model.txt", "r" )
```

```python
cols = df3.readline().strip().split(",")
print(cols)
n_cols = len(cols)
```

```python
n_cols
```

```python
count = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in df3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    count += 1

print(f'Total of number of rows is {count}, number of columns {n_cols} ')
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

### Read and Write a File with Python

```python
intFile = "../datasets/customer-churn-model/Customer Churn Model.txt"
outFile = "../datasets/customer-churn-model/Tab2 Customer Churn Model.txt"
```

```python
with open(intFile, "r") as intFile1:
    with open(outFile, "w") as outFile1:
        for line in intFile1:
            fields = line.strip().split(',') 
            outFile1.write("\t".join(fields))
            outFile1.write("\n")
```

```python
df4 = pd.read_csv(outFile, sep='\t')
df4.head()
```
